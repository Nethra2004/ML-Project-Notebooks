IMPORTING THE DEPENDENCIES

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

DATA LOADING AND PREPROCESSING

```python
movie_metadata = pd.read_csv('/content/movies_metadata.csv.zip', low_memory=False)
```

```python
movie_metadata = movie_metadata[['title', 'genres', 'popularity', 'vote_average', 'vote_count']].dropna()
```

NORMALISING NUMERIC COLUMNS

```python
scaler = StandardScaler()
```

```python
movie_metadata[['popularity', 'vote_average', 'vote_count']] = scaler.fit_transform(movie_metadata[['popularity', 'vote_average', 'vote_count']])
```

INSTANTIATING AND TRAINING KNN CLASSIFIER

```python
knn_classifier = NearestNeighbors(metric='cosine', algorithm='brute')
knn_classifier.fit(movie_metadata[['popularity', 'vote_average', 'vote_count']])
```

RECOMMENDING SIMILAR MOVIES

```python
def recommend_similar_movies(movie_index, k=5):
    distances, indices = knn_classifier.kneighbors(movie_metadata[['popularity', 'vote_average', 'vote_count']].iloc[movie_index, :].values.reshape(1, -1), n_neighbors=k+1)
    similar_movies = []
    for i in range(1, len(distances.flatten())):
        similar_movies.append(movie_metadata.iloc[indices.flatten()[i], 0])
    return similar_movies
```

```python
target_movie_index = 0
similar_movies = recommend_similar_movies(target_movie_index)
print(f"Recommended Similar Movies for '{movie_metadata.iloc[target_movie_index, 0]}':")
for movie_title in similar_movies:
    print(movie_title)
```

PREPARING THE DATA

```python
X = movie_metadata[['popularity', 'vote_average', 'vote_count']]
y = movie_metadata['genres']
```

TRAIN/TEST SPLIT

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

INSTANTIATING AND TRAINING KNN CLASSIFIER

```python
knn_classifier = NearestNeighbors(metric='cosine', algorithm='brute')
knn_classifier.fit(X_train)
```

PREDICTING LABELS

```python
distances, indices = knn_classifier.kneighbors(X_test)
```

EVALUATING THE ACCURACY

```python
def evaluate_accuracy(predictions, actual_labels, indices):
    predicted_labels = []
    for neighbor_indices in indices:
        neighbor_labels = y_train.iloc[neighbor_indices].mode()[0]
        predicted_labels.append(neighbor_labels)
    accuracy = accuracy_score(actual_labels, predicted_labels)
    return accuracy
accuracy = evaluate_accuracy(y_test, y_test, indices)
print("Accuracy:", accuracy)
```
