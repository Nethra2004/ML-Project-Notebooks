IMPLEMENTATION OF LINEAR REGRESSION AND GRADIENT DESCENT FOR MINIMUM COST CALCULATION

IMPORTING NECESSARY LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

LOADING THE DATASET

```python
dataset = pd.read_csv('/content/archive.zip')
```

DEFINING THE COST FUNCTION(MEAN SQUARED ERROR)

```python
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

GRADIENT DESCENT OPTIMISATION

```python
def gradient_descent(X, y, learning_rate, iterations):
    m = 0
    b = 0
    n = len(X)
    for _ in range(iterations):
        y_pred = m*X + b
        dm = (-2/n) * np.sum(X * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)
        m -= learning_rate * dm
        b -= learning_rate * db

    return m, b
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 3 + 4 * X + np.random.randn(100, 1)
learning_rate = 0.01
iterations = 1000
```

PERFORMING GRADIENT DESCENT

```python
m, b = gradient_descent(X, y, learning_rate, iterations)
```

PLOTTING THE REGRESSION LINE

```python
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, m*X + b, color='red', label='Regression Line')
plt.title('Linear Regression with Gradient Descent')
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.legend()
plt.show()
```

CALCULATING THE MINIMUM COST

```python
y_pred = m*X + b
cost = mean_squared_error(y, y_pred)
print("Minimum Cost achieved:", cost)
```
